```python
import pandas as pd

df = pd.read_csv('Iris.csv')
df
```

```python
df['Species'].unique()
```

```python
Iris_setosa = df[df['Species'] == 'Iris-setosa']
Iris_versicolor = df[df['Species'] == 'Iris-versicolor']
Iris_virginica = df[df['Species'] == 'Iris-virginica']
```

```python
print("Iris-setosa shape: ", Iris_setosa.shape)
print("Iris-versicolor shape:", Iris_versicolor.shape)
print("Iris-virginica shape:", Iris_virginica.shape)
```

```python
merged = pd.concat([Iris_setosa, Iris_versicolor])
print(merged)
```

```python
print(merged.shape)
```

```python
sorted_values = df.sort_values(by='PetalLengthCm', ascending=False)
print(sorted_values)
```

```python
melted1 = pd.melt(df, id_vars=['Species', 'Id'], value_vars=['SepalLengthCm','SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'], var_name='Metrics', value_name='Value')
print(melted1)
```

```python
wide1 = melted1.pivot_table(index=['Species', 'Id'], columns='Metrics' ,values='Value', aggfunc="sum")
```
