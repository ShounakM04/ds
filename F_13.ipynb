```python
import pandas as pd
import numpy as np

df = pd.read_csv('heart.csv')
df.shape
```

```python
df.isna().sum()
```

```python
df.isnull().sum()
```

```python
num_cols = df.select_dtypes(include=np.number).columns

num_cols
```

```python
for col in num_cols:
  df[df[col] < 0] = np.nan
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3-Q1
lower = Q1-1.5*IQR
higher = Q3+1.5*IQR

for col in df.columns:
  df = df[(df[col] >= lower[col]) & (df[col] <= higher[col])]

print(df.shape)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

numscold = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'ca']

df_scaled = df.copy()

scaled = sc.fit_transform(df[numscold])

df_scaled[numscold] = scaled

print(df_scaled)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(df_scaled.drop('target', axis=1), df_scaled['target'], test_size=0.2, random_state=42)

lr = LogisticRegression()

lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
knn = KNeighborsClassifier()

knn.fit(X_train, y_train)

y_pred1 = knn.predict(X_test)

accuracy_score(y_test, y_pred1)
```
