```python
melted = pd.melt(df, id_vars=['workclass', 'education', 'occupation'], value_vars=['capital-gain', 'capital-loss', 'hours-per-week', 'income'], var_name="Income Stats", value_name="Value")
print(melted)
```

```python
df = df[(df['workclass'] != '?') & (df['occupation'] != '?') & (df['native-country'] != '?')]

melted1 = pd.melt(
    df,
    id_vars=['gender', 'education', 'workclass'],
    value_vars=['capital-gain', 'capital-loss', 'hours-per-week'],
    var_name='Income Feature',
    value_name='Value'
)
print(melted1)
```

```python
cast = melted1.pivot_table(index=["gender", "education", "workclass"], columns="Income Feature", values="Value", aggfunc="sum")
print(cast)
```
