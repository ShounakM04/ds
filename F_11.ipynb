```python
melted = pd.melt(df, id_vars=["genres"], value_vars=["vote_average", "vote_count"], var_name="vote metrics", value_name="Value")
print(melted)
```

```python
cast = melted.pivot_table(index=["genres"], columns=["vote metrics"], values="Value", aggfunc="mean")
print(cast)
```
