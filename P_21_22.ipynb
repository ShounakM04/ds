```python
import pandas as pd
df = pd.read_csv('adult.csv')
df
```

```python
df.hist(figsize=(10,12), bins=20, edgecolor="black", grid=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.histplot(data=df['hours-per-week'], bins=20, edgecolor="black")
plt.xlabel('hours-per-week')
plt.ylabel('count')
plt.title('Hours per week')
plt.show()
```

```python
sns.histplot(data=df['age'], bins=20, edgecolor="black")
plt.xlabel('age')
plt.ylabel('count')
plt.title('Age of people')
plt.show()
```

```python
sns.stripplot(data=df, x='relationship', y='age', jitter=True, alpha=0.6)
```

```python
sns.stripplot(data=df, x='income', y='age', jitter=True, alpha=0.5)
plt.title('Age distribution by Income Group')
plt.xlabel('Income')
plt.ylabel('Age')
plt.show()
```

```python
sns.barplot(data=df, x='income', y='age', hue='gender')
```

```python
avgdata = df.groupby('age')['hours-per-week'].mean().reset_index()
avgdata
```

```python
sns.lineplot(data=avgdata, x='age', y='hours-per-week')
```

```python
df['gender'].value_counts().plot.pie(autopct='%1.2f%%')
```
